## Explorando Datasets - Clientes

```
# Conhecendo o dataset
df_clientes.head(3)
```

```
# Quais cidades tem no meu dataset?
print(df_clientes.customer_city.unique())
```

```
# Quantas cidades são?
len(df_clientes.customer_city.unique())
```

```
# Total de clientes
len(df_clientes.customer_unique_id.unique())
```

```
# Compreendendo melhor a relação entre customer_unique_id e customer_id

df_group = df_clientes.groupby( by = 'customer_unique_id' ).count()
df_group.head(10)
```

```
# Filtrando apenas os que chamam atenção (contagem de customer_id >= 4)

df_group[ (df_group['customer_id'] >= 4) ].head(5)
```

Observa-se que um mesmo cliente possui um customer_id para cada compra realizada.

```
# Quais as 5 cidades com maior quantidade de compras?

# Usando COUNT + SORT VALUES...
df_clientes.groupby(by= 'customer_city').count().sort_values(by='customer_id', ascending=False).head(5)
```

```
# ...ou simplificando com VALUE COUNTS...
df_top5_compras = df_clientes.customer_city.value_counts().head(5)
df_top5_compras
```

```
# ... ou graficamente
data = df_top5_compras.reset_index()

fig = px.bar(data, x='index', y='customer_city',
             labels= {'index': 'Cidades', 'customer_city': 'Qtd de Compras'},
             title= 'Top 5 - Cidades com mais compras')

gitly.show( fig )
```
